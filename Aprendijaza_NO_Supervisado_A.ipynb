```python
#Corre el algoritmo limpio del inicio, presentar las predicciones y la inercia del mismo.
k=4
kmeans=KMeans(n_clusters=k)
y_pred=kmeans.fit_predict(x)
kmeans.inertia_
```

```python
# Toma el algoritmo del inicio para calcular las inercias cuando se utiliza de 1 a 9 clusters
x2 = np.linspace(1,8,8)
y2 = []
for i in range(1,9):
    k = i
    kmeans = KMeans(n_clusters=k)
    y_pred = kmeans.fit_predict(x)
    a = kmeans.inertia_
    y2.append(a)
```

```python
#Grafica las inercias anteriores
plt.plot(x2,y2)
plt.xlabel('Cantidad de cluster')
plt.ylabel('Inercia')
plt.title('Comparación de la inercia vs numero de Cluster')
plt.grid()
plt.show()
```

```python
#Obtenen el coeficiente de silueta
from sklearn.metrics import silhouette_score, silhouette_samples
silhouette_score(x, kmeans.labels_)
```

```python
#Ejecuta el algoritmo varias veces con diferentes clusters y obtenen sus coeficientes de silueta
x2 = np.linspace(2,8,7)
y2 = []
for i in range(2,9):
    k = i
    kmeans2 = KMeans(n_clusters=k)
    kmeans2.fit_predict(x)
    a = silhouette_score(x, kmeans2.labels_)
    y2.append(a)
```

```python
#Grafica los coeficientes de silueta
plt.plot(x2,y2)
plt.xlabel('Cantidad de cluster')
plt.ylabel('Coeficiente de Silueta')
plt.title('Comparación del coef. Silueta vs Numero de Cluster')
plt.grid()
plt.show()
```

```python
#Realiza el diagrama de silueta
import matplotlib.cm as cm
X = x
range_n_clusters = [2, 3, 4, 5, 6]

for n_clusters in range_n_clusters:

    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    ax1.set_xlim([-0.1, 1])
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])


    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)


    silhouette_avg = silhouette_score(X, cluster_labels)
    print(
        "Para n_clusters =",
        n_clusters,
        "La media de silhouette_score es :",
        silhouette_avg,
    )

    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            ith_cluster_silhouette_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7,
        )

        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        y_lower = y_upper + 10

    ax1.set_title("La grafica de siluetas para varios clusters")
    ax1.set_xlabel("Los valores del coeficiente de siluetas")
    ax1.set_ylabel("Etiqueta del cluster")

    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(
        X[:, 0], X[:, 1], marker=".", s=30, lw=0, alpha=0.7, c=colors, edgecolor="k"
    )

    centers = clusterer.cluster_centers_

    ax2.scatter(
        centers[:, 0],
        centers[:, 1],
        marker="o",
        c="white",
        alpha=1,
        s=200,
        edgecolor="k",
    )

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker="$%d$" % i, alpha=1, s=50, edgecolor="k")

    ax2.set_title("La grafica de los puntos con su cluster.")

    plt.suptitle(
        "Analisis de silueta para clustering de datos con n_clusters = %d"
        % n_clusters,
        fontsize=14,
        fontweight="bold",
    )

plt.show()
```

#### <span style="color:blue"> 1.4 Clustering con Preprocesamiento</span>

```python
#Importa los dígitos de MNIST desde Sklearn
from sklearn.datasets import load_digits
from sklearn import datasets
#Asigna las variables x, y
x,y = load_digits(return_X_y=True)
digits = datasets.load_digits()
```

```python
print(x)
print('----------------------')
print(y)
```

```python
fig, ax = plt.subplots(8, 8, figsize=(6,6))
for i, axi in enumerate(ax.flat):
    axi.imshow(digits.images[i], cmap='binary')
    axi.set(xticks=[], yticks=[])
plt.show()
```

```python
#Importa train_test_split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
#Arma una regresión logística
from sklearn.linear_model import LogisticRegression

log = LogisticRegression()
log.fit(x_train, y_train)
```

```python
#Despliega el puntaje de la regresión (log.score)
log.score(x_test, y_test)
```

```python
#Arma un pipeline que incluya el algoritmo de K-Means
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ("kmeans", KMeans(n_clusters=30)),
    ("log", LogisticRegression())
])
pipeline.fit(x_train, y_train)
```

```python
#Despliega el puntaje del pipeline anterior
pipeline.score(x_test, y_test)
```

```python
#Arma un GridSearch para identificar con cuántos clusters se puede correr
from sklearn.model_selection import GridSearchCV

param_grid = dict(kmeans__n_clusters=range(2, 100))
grid = GridSearchCV(pipeline, param_grid, cv=5, verbose=2)
grid.fit(x_train, y_train)
```

```python
#Obtene el mejor parámetro con grid.best_params_
grid.best_params_
```

```python
#Obtene el puntaje del GridSearch (Grid.score)
grid.score(x_test, y_test)
```

#### <span style="color:blue"> 1.4 DBSCAN</span>

El algoritmo **DBSCAN** es un método de agrupamiento que se basa en la densidad de los datos en el espacio.

A diferencia de otros algoritmos de agrupamiento como k-means, DBSCAN no requiere especificar de antemano el número de clústeres. En su lugar, determina automáticamente el número de clústeres basándose en la distribución de los datos.

El algoritmo DBSCAN utiliza dos parámetros principales: el radio de búsqueda (epsilon, ε) y el número mínimo de puntos (MinPts) requeridos para formar un clúster. Estos parámetros influyen en cómo se agrupan los puntos en función de su densidad.

El proceso de DBSCAN es el siguiente:

1. Seleccionar un punto de datos aleatorio que no haya sido visitado anteriormente.
2. Encontrar todos los puntos que están dentro de una distancia ε del punto seleccionado (incluido él mismo). Estos puntos forman el "vecindario" del punto inicial.
3. Si el número de puntos en el vecindario es menor que MinPts, el punto se marca como ruido o "outlier".
4. Si el número de puntos en el vecindario es igual o mayor que MinPts, se crea un nuevo clúster y se expande añadiendo puntos adicionales. Estos puntos deben estar en el vecindario del punto inicial o en el vecindario de otros puntos ya incluidos en el clúster.
5. Repetir los pasos anteriores hasta que se hayan visitado todos los puntos.
6. Cualquier punto que no haya sido asignado a un clúster se considera ruido.

Puede funcionar cuando K-Means esta limitado. Este algoritmo define los clusters como regiones continuas de alta densidad.

```python
#Crea set de datos con make_moons y graficarlo
from sklearn.datasets import make_moons

x,y = make_moons(n_samples=500, noise=0.1)
plt.scatter(x[:,0],x[:,1])
plt.title('Datos en forma de medias lunas')
plt.grid()
plt.show()
```

```python
#Importa DBSCAN de Sklearn y correrlo
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.05, min_samples=5)
dbscan.fit(x)
```

```python
#Verifica como se hizo la agrupación de las instancias con la función labels_
dbscan.labels_
```

```python
#Modifica el parámetro epsilon del algoritmo DBSCAN
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.2, min_samples=10)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
#Importa Kneighbors Classifier para clasificar los datos de DBSCAN
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=50)
knn.fit(dbscan.components_, dbscan.labels_[dbscan.core_sample_indices_])
```

```python
#predice las clasificación de datos aleatorios para ejemplificar el método
x_nuevo = np.array([[-0.5, 0], [0, 0.5], [1, -0.1], [2, 1]])
knn.predict(x_nuevo)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

# Generar datos de ejemplo (dos lunas)
X, y = make_moons(n_samples=200, noise=0.05)

# Crear el objeto DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=10)

# Ajustar el modelo DBSCAN a los datos
dbscan.fit(X)

# Obtener las etiquetas de los clústeres
labels = dbscan.labels_

# Obtener el número de clústeres en el resultado (ignorando el ruido)
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)

# Imprimir los resultados
print("Número de clústeres encontrados:", n_clusters_)
print("Etiquetas de los clústeres:", labels)

# Graficar los puntos y colorear según las etiquetas de los clústeres
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("DBSCAN Clustering")
plt.grid()
plt.show()

```
