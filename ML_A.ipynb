```python
#Se importan librerias
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Se carga el data Set
iris=load_iris()
X=iris.data
y=iris.target
```

```python
print(list(iris.target_names))
```

```python
#Datos de entrada
print(X)
```

```python
#La salida
print(y)
```

```python
#El tama√±o de la salida
print(len(y))
```

```python
#Dividir el data set en conjunto de datos de prueba y de entrenamaiento
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.1,random_state=123, stratify=y)
print(X_train)
print('--------------------------')
print(y_test)
```

```python
#Datos de entrada para probar
print(X_test)
```

```python
#Datos de prueba de salida
print(y_test)
```

```python
#Implementa el modelo de los vecinos cercanos
Knn=KNeighborsClassifier(n_neighbors=3)
Knn.fit(X_train,y_train)
print('------------------------')
print('Resultados obtenidos')
print(Knn.predict(X_test))
print('------------------------')
print('Score de validaci√≥n')
Knn.score(X_test,y_test)
```

```python
#Nuevas observaciones
observacion=[[5.3,2,4.1,1.8],[7,2.8,4.9,2.],[7.7,4,6.7,2.],[6.3,2.7,3,1.8]]
Knn.predict(observacion)
```

```python
#Prueba de numero de vecino
train_scores={}
test_scores={}
neighbors=np.arange(1,10)
for neighbor in neighbors:
  Knn=KNeighborsClassifier(n_neighbors=neighbor)
  Knn.fit(X_train,y_train)
  train_scores[neighbor]=Knn.score(X_train,y_train)
  test_scores[neighbor]=Knn.score(X_test,y_test)

print('------------------')
print('Resultado de Score de entrenamiento')
print(train_scores)
print('------------------')
print('Resultado de Score de Prueba')
print(test_scores)
```

```python
#Grafica de los puntajes
plt.plot(list(train_scores.values()), 'r--', label='Train')
plt.plot(list(test_scores.values()), 'b*-', label='Test')
plt.legend()
plt.xlabel('Vecios')
plt.ylabel('Porcentaje')
plt.title('Comparaci√≥n de rendimiento de vecinos')
plt.grid()
plt.show()
```

```python
#Revisi√≥n de la clasificaci√≥n
print(y_test)
print(Knn.predict(X_test))
```

```python
#Mariz de confusi√≥n
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,Knn.predict(X_test))
```

```python
#Algoritmo de agrupamiento K-MEANS
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
```

```python
iris=datasets.load_iris().data
#print(iris)
```

```python
inercias=[]
for i in range(1,len(iris+1)):
  Kmeans=KMeans(n_clusters=i,max_iter=300)
  Kmeans.fit(iris)
  inercias.append(Kmeans.inertia_)

```

```python
plt.plot(range(1,len(iris+1)),inercias)
plt.xlabel('Clusters')
plt.ylabel('Inercia')
plt.title('Iteraciones vs Inercia')
plt.grid()
plt.show()
```

```python
#Zoom a la grafica
plt.plot(range(1,21),inercias[:20])
plt.xlabel('Clusters')
plt.ylabel('Inercia')
plt.title('Iteraciones vs Inercia')
plt.grid()
plt.show()
```

```python
#Zoom a la grafica
plt.plot(range(1,11),inercias[:10])
plt.xlabel('Clusters')
plt.ylabel('Inercia')
plt.title('Iteraciones vs Inercia')
plt.grid()
plt.show()
```

```python

#Modelo de agrupamiento
Kmeans=KMeans(n_clusters=3, max_iter=300)
objetivos=Kmeans.fit_predict(iris)
print(objetivos)
print('----------------------')
print(datasets.load_iris().target)
```

```python
#Plot de los grupos
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.set_xlabel('Ancho del sepalo')
ax.scatter(iris[objetivos==0,0],iris[objetivos==0,3],iris[objetivos==0,2],
           s=iris[objetivos==0,1],color='green')

ax.scatter(iris[objetivos==1,0],iris[objetivos==1,3],iris[objetivos==1,2],
           s=iris[objetivos==1,1],color='red', marker='o')

ax.scatter(iris[objetivos==2,0],iris[objetivos==2,3],iris[objetivos==2,2],
           s=iris[objetivos==2,1],color='blue', marker='s')

ax.scatter(Kmeans.cluster_centers_[:,0], Kmeans.cluster_centers_[:,3],
          Kmeans.cluster_centers_[:,2], s=35, color='black', marker='X')

plt.show()
```

```python
#Redes Neuronales
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import  OneHotEncoder

from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
#cargar los datos
iris=load_iris()
X=iris.data
y_1=iris.target.reshape(-1,1)
```

```python
print(y_1)
```

```python
#Se hace la codificaci√≥n
encoder=OneHotEncoder(sparse=False)
y_1=encoder.fit_transform(y_1)
X_train,X_test,y_train,y_test=train_test_split(X,y_1,test_size=0.1,random_state=5, stratify=y_1)
```

```python
print(y_1)
```

```python
print(len(y_1))
print(len(X_test))
print(len(X_train))
print(len(y_train))
print(len(y_test))
```

```python
#Construye el modelo de la red Neuronal
model=Sequential()
model.add(Dense(10, input_shape=(4,), activation='relu')) #Capa 2 oculta
model.add(Dense(10, activation='relu')) #Capa 2  oculata
model.add(Dense(3, activation='softmax')) # Capa de salida
```

```python
#Compilaci√≥n del modelo
optimizer=Adam(lr=0.001)
model.compile(optimizer,loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
#Entrenamiento del modelo
model.fit(X_train, y_train, epochs=100)
results=model.evaluate(X_test, y_test)


print('----------------------------------------------------')
print('Pedida final de la prueba: {:4f}'.format(results[0]))
print('Precision final de la prueba: {:4f}'.format(results[1]))
```
